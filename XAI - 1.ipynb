# Line Plot

A line plot, the most basic technique, can be used for plotting the relationship or dependence of a variable on another variable. To do so, we just have to call the function of the plot.

In our example, we have taken two arrays X1 and Y1 which contains points on X-axis and points on Y-axis respectively for first dataset.

Line Plot is usually used when the points increase uniformly to represent the data.

```
import matplotlib.pyplot as plt     #For representation
import numpy as np                  #For calculations with array
x1 = [1,3,5,7,9]                    #To be plotted on X-axis for first dataset
y1 = [2,3,4,5,6]                    #To be plotted on Y-axis for first dataset
x2 = [1,4,6,8,12,17]                #To be plotted on X-axis for second dataset
y2 = [2,4,5,6,7,8]                  #To be plotted on Y-axis for first dataset
plt.plot(x1,y1, label = "this is first")      #plot is used to create the graph
plt.plot(x2,y2, label = "this is second")
plt.title("LINE PLOT")
plt.legend()
plt.xlabel("this is X label")
plt.ylabel("this is y label")
plt.show()                          #show is used to display the graph

```

# Bar Chart
Bar charts are handy to compare the quantitative values of various groups. They can be shown in the form of bars, either vertical or horizontal. The length/height of the bar corresponds to its value.

In our example, we have taken two datas to compare.

when trying to measure change over time, bar graphs are best when the changes are larger.






```
plt.bar(x1,y1,label="this is Bar plot for first Dataset",color = 'g')  
plt.bar(x2,y2,label="this is Bar plot for second Data-set",color='r')
plt.title("this is bar plot")
plt.xlabel("this is X label")
plt.ylabel("this is y label")
plt.legend()
plt.show()
```

# Pie and Donut Charts
Pie and donut charts are good for comparison of the parts of a whole. They are particularly effective when the components are limited with text and percentages also describing the content. At the same time, it is not easy to interpret them as we struggle to estimate areas and compare different visual angles.




```
slices = [84,96,94,90]                              #Parts of the whole
activities = ["sem1","sem2","sem3","sem4"]          #Labels
plt.pie(slices,labels=activities,explode=(.1,.1,.1,.1),shadow = True,startangle = 20,autopct='%1.1f%%')     #Properties
```

# Histogram Plot
A histogram depicts how a continuous variable is distributed through an interval or time frame. It is a frequently used data visualization technique in the area of machine learning. The data are plotted by lumping them into intervals which are known as ‘bins’. It reveals how the frequency is distributed and where the deviations and asymmetries are. 


```
population_ages = [22,55,62,45,21,22,34,42,42,4,99,102,110,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48]           #Data to be plotted
plt.hist(population_ages,bins=15,rwidth = 1)                                                                              #hist is used to make graph
```

# Scatter Plot 
A scatter plot or scattergram chart will show the relationship between two different variables or it can reveal the distribution trends. 

It should be used when there are many different data points, and you want to highlight similarities in the data set. This is useful when looking for outliers or for understanding the distribution of your data.

```
plt.scatter(x1,y1,label="this is first data set",color='g',s=25,marker="o")     #plt.scatter is used to build scatter graph
plt.scatter(x2,y2,label="this is second data set",color='r',s=25,marker="*")
```

#Box and whisker plot
Box and whisker plot is used when you have multiple data sets from independent sources that are related to each other in some way and to display the variation in set of data.A box and whisker plot can provide additional detail while allowing multiple sets of data to be displayed in the same graph.

Here I have given the example in which data of 12 months production of product units of 4 manufacturing compnies is given.The data is shown in Box and whisker plot.The plot shows the variation between the production of 4 companies.

```
import matplotlib.pyplot as plt
%matplotlib inline

company1=[4000,3500,5700,4800,5900,4200,5000,4900,4100,6000,5200,4500]
company2=[5000,5900,6000,4900,3900,5900,5700,4000,4200,3500,4100,4500]
company3=[4000,4500,3900,5000,4200,4200,5200,4700,5900,6000,5100,4950]
company4=[4700,4500,4500,5000,4400,4800,4400,3900,4000,3900,5100,4800]

plt.boxplot([company1,company2,company3,company4])
```

Here from the boxplot we can see that company2 is having highest variation than other companies and company4 is having least variation for the production of units in 12 months.The range of the highest and lowest of production unit of individual company is also shown.The median(average of middle two values) unit production of individual company is shown by horizontal yellow line in the plot.

**Network diagram:**

Network diagrams visualize interconnections and relationships. They highlight interconnections between entities through the use of nodes and lines.There are two types of network diagrams, directed and undirected.Network diagrams can provide insights regarding structure and cluster patterns of network and can help identify the problem if there is a breakdown somewhere in the network.

We can take an example of route system in between different cities.The routes can be considered as lines and cities can be considered as nodes.The network diagram of this system can visualize the interconnection between cities.

```
import networkx as nx
import matplotlib.pyplot as plt
g=nx.Graph()
g.add_node("a")    #here a,b,c,d,e,f,g,h,i,j,k are cities
g.add_node("b")
g.add_node("c")
g.add_node("d")
g.add_node("e")
g.add_node("f")
g.add_node("g")
g.add_node("h")
g.add_node("i")
g.add_node("j")
g.add_node("k")

g.add_edge("a","b")
g.add_edge("a","e")
g.add_edge("f","b")
g.add_edge("k","e")
g.add_edge("b","d")
g.add_edge("g","d")
g.add_edge("d","c")
g.add_edge("a","b")
g.add_edge("j","b")
g.add_edge("f","i")
g.add_edge("g","k")
g.add_edge("h","b")
g.add_edge("h","k")
g.add_edge("d","i")

print(nx.info(g))
nx.draw(g,with_labels=True)
plt.show()
```

**Correlation Matrices:**

A correlation matrix is a table showing correlation coefficients between sets of variables. the matrix shows that each random variable in the table is correlated with each of the other values in the table. This allows you to see which pairs have the highest correlation.

Here I have taken an example of various factors affecting weather such as temperature,humidity,wind speed etc and shown correlating between them using Correlation matrix.

```
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import numpy as np
 
data={                                                      #Dictionary containing variables and there data
    'Temperature':[32,37,30,40,45,39,35,29,42],           
    'Humidity(in percentage)':[30,15,35,20,17,10,18,25,40],
    'dryness(in percentage)':[7,25,5,15,20,30,20,13,2],
    'Air Pressure':[45,36,30,40,29,31,38,42,35],
    'Density':[30,26,34,40,29,38,42,31,39],
    'Cloud Cover(in percentage)':[30,10,35,15,12,5,13,25,40],
    'wind speed':[45,34,29,30,35,20,37,40,50],
}
 
 
df=pd.DataFrame(data)
 
mask=np.zeros_like(df.corr())
triangle_indices=np.triu_indices_from(mask)
mask[triangle_indices]=True
 
plt.figure(figsize=(7,7))
seaborn.heatmap(df.corr(),mask=mask,annot=True,annot_kws={"size":12})
seaborn.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
